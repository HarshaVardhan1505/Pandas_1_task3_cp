#**What is Pandas?**

*   **Pandas Library** in python is like a spreadsheet software, such as excel or Googlesheets, but it offers greater efficiency and functionality for **tabular data handling , manipulations** and **more customizable operations**.

*   Here we can write **own formulas** and **own logics**.
*   The pandas library is more useful for the data analyst,data science etc.

*   Here we mainly do **data analysis,data manipulation,data visualization** but visualization is not mostly used but helps a lot.
* It is an **open source** and **invented** in the **year** of **2004**.
* It is **actively developing** and having the **greater Community**.






##**The Highlights Of Pandas**

* It is fast and efficient data frame object for Data manipulation with integrated indexing.
* The tools for reading and writing data between in memory data structures and different formats : **CSV Files**,**Text Files**,**TSV Files**,**Microsoft Excel**,**SQL Databases** and **Fast HDF5 Format**.
* Columns can be inserted and deleted from data structures for size mutability.


## **Why Use Pandas ?**

* Pandas allows to **analyze big data** and make **conclusions** based on **statistical theories**.
* Pandas can **clean messy datasets** and make them **readable** and **relevant**.
* **Relevant** data is very **important** in data science.

##**What Can Pandas Do ?**

* Pandas gives us the **answers** about the data like :
 * Is there a **correlation between two** or **more columns** ?
 * What is **average value**?
 * **Maximum value** ?
 * **Minimum value** ?
* Pandas are also to delete rows that are not relevant or contains wrong values like **Empty** or **Null values**.

##**What We Can Do With Pandas ?**

* We can make **data loading** and **reading** it means we can easily import data from various **file formats** like **CSV** , **TSV** ,**Excel** and **Databases**.
* We can make **data cleaning** and **manipulations** it means **cleaning** and preparing a data for analysis by **handling missing values** , **duplicates** and **data type conversions**.
* We can make **data exploration** and **analysis** if means exploring a data using **Summary Statistics** ,**Filtering, Sorting,Grouping** and **Aggregate Functions.**
* We can make **Data Visualization** it means we can **create insight full data visualization**
